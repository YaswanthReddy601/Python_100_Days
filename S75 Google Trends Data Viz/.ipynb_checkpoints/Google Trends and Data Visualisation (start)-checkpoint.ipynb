### Tesla Stock Price v.s. Search Volume

**Challenge:** Plot the Tesla stock price against the Tesla search volume using a line chart and two different axes. Label one axis 'TSLA Stock Price' and the other 'Search Trend'. 

```python
import matplotlib.pyplot as mpl

y1= mpl.gca()
y2= mpl.twinx()

y1.set_ylabel("Tesla Stock Price")
y2.set_ylabel("Tesla Web Search")


y1.plot(df_tesla.MONTH, df_tesla.TSLA_USD_CLOSE)
y2.plot(df_tesla.MONTH, df_tesla.TSLA_WEB_SEARCH)
```

**Challenge**: Add colours to style the chart. This will help differentiate the two lines and the axis labels. Try using one of the blue [colour names](https://matplotlib.org/3.1.1/gallery/color/named_colors.html) for the search volume and a HEX code for a red colour for the stock price. 
<br>
<br>
Hint: you can colour both the [axis labels](https://matplotlib.org/3.3.2/api/text_api.html#matplotlib.text.Text) and the [lines](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D) on the chart using keyword arguments (kwargs).  

```python

y1= mpl.gca()
y2= mpl.twinx()

y1.set_ylabel("Tesla Stock Price", color="#FF7333")
y2.set_ylabel("Tesla Web Search", color="#44686B")

y1.plot(df_tesla.MONTH, df_tesla.TSLA_USD_CLOSE, "#FF7333")
y2.plot(df_tesla.MONTH, df_tesla.TSLA_WEB_SEARCH, "#44686B")
```

**Challenge**: Make the chart larger and easier to read. 
1. Increase the figure size (e.g., to 14 by 8). 
2. Increase the font sizes for the labels and the ticks on the x-axis to 14. 
3. Rotate the text on the x-axis by 45 degrees. 
4. Make the lines on the chart thicker. 
5. Add a title that reads 'Tesla Web Search vs Price'
6. Keep the chart looking sharp by changing the dots-per-inch or [DPI value](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.figure.html). 
7. Set minimum and maximum values for the y and x axis. Hint: check out methods like [set_xlim()](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.set_xlim.html). 
8. Finally use [plt.show()](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.show.html) to display the chart below the cell instead of relying on the automatic notebook output.

```python
#setting fig size
mpl.figure(figsize=(14,8), dpi=120)

ax1=mpl.gca()#gets the axis
ax2=mpl.twinx()#crets a twin axis

#setting the labels on both y axis
ax1.set_ylabel("Tesla Stock Price", fontsize=14, color="#FF7333")
ax2.set_ylabel("Tesla Web Search", fontsize=14, color="#44686B")
mpl.xticks(fontsize=14, rotation=45)

#giving title to the figure
mpl.title("Tesla Stock Price v.s. Search Volume", fontsize=20, color="green")

#setting axis limits
ax1.set_ylim(0,500)
ax2.set_ylim(0,30)

#ploting the graphs
ax1.plot(df_tesla.MONTH, df_tesla.TSLA_USD_CLOSE, "#FF7333")
ax2.plot(df_tesla.MONTH, df_tesla.TSLA_WEB_SEARCH, "#44686B")

#displaying the graph
mpl.show()
```

### Including 2020 in Unemployment Charts

**Challenge**: Read the data in the 'UE Benefits Search vs UE Rate 2004-20.csv' into a DataFrame. Convert the MONTH column to Pandas Datetime objects and then plot the chart. What do you see?

```python
import pandas
```

```python
df_ue = pandas.read_csv("UE Benefits Search vs UE Rate 2004-20.csv")
```

```python
df_ue.MONTH= pandas.to_datetime(df_ue.MONTH)
```

```python
#setting fig size
mpl.figure(figsize=(14,8), dpi=120)

ax1=mpl.gca()#gets the axis
ax2=mpl.twinx()#crets a twin axis

#setting the labels on both y axis
ax1.set_ylabel("FRED U/E Rate", fontsize=14, color="#FF7333")
ax2.set_ylabel("BTC Web Search", fontsize=14, color="#44686B")
mpl.xticks(fontsize=14, rotation=45)

#giving title to the figure
mpl.title("'Unemployment Benefits' in the U.S. vs the U/E Rate", fontsize=20, color="green")

#setting axis limits
# ax1.set_ylim(0,100)
# ax2.set_ylim(0,16000)
# ax1.grid()

#Creating the ticks
import matplotlib.dates as mdates

years= mdates.YearLocator()
months= mdates.MonthLocator()
years_fmtr= mdates.DateFormatter("%Y")

ax1.xaxis.set_major_locator(years )
ax1.xaxis.set_major_formatter(years_fmtr)
ax1.xaxis.set_minor_locator(months )


#ploting the graphs
ax1.plot(df_ue.MONTH, df_ue.UE_BENEFITS_WEB_SEARCH, "#FF7333")
ax2.plot(df_ue.MONTH, df_ue.UNRATE, "#44686B")

#displaying the graph
mpl.show()
```
